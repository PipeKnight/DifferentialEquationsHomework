a) Строим поле направлений для уравнения:
$$ \dot x = -\frac{2t}{5x} $$

```python
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'svg'
import numpy as np
import odebook as ob
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("$t$", fontsize=20)
plt.ylabel("$x$", fontsize=20)
deltax = 1
border = 10 * deltax + deltax / 2
t = np.mgrid[-border:border:deltax]
x = np.mgrid[-border:border:deltax]
ob.normdirfield(t, x, (lambda t, x: -2 * t / 5 / x), length=deltax*0.9)
```

b) Решаем диф. уравнение:
$$
\frac{dx}{dt} = \frac{-2t}{5x} \\
5x dx = -2t dt \\
\int 5x dx = \int -2t dt \\
2.5 x^2 = -t^2 + C \\
x(t) = \pm \sqrt{\frac{2}{5} (C - t^2)}
$$

c) Находим частные решения:

$ x(0) = 2 \Rightarrow C = 10 $ и решение имеет вид: $ x(t) = \sqrt{\frac{2}{5} (10 - t^2)} $

$ x(0) = -1 \Rightarrow C = \frac{5}{2}$ и решение имеет вид: $ x(t) = -\sqrt{\frac{2}{5} (2.5 - t^2)} $
