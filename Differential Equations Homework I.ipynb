b) Решаем диф. уравнение:
$$
\frac{dx}{dt} = \frac{-x}{4t} \\
-\frac{dx}{x} = \frac{dt}{4t} \\
\int -\frac{dx}{x} = \int \frac{dt}{4t} \\
- \ln{|x|} = \frac{1}{4} \ln{|t|} + C \\
|x| = \frac{1}{\sqrt[4]{\pm t}} \cdot e^c \\
x(t) = \frac{C}{\sqrt[4]{\pm t}}
$$

Область определения $(0; +\infty)$

$ x(-1) = 1 \Rightarrow C = 1 $ и решение имеет вид: $ x(t) = \frac{1}{\sqrt[4]{-t}} $.
Область определения $(-\infty; 0)$

$ x(-1) = -1 \Rightarrow C = -1 $ и решение имеет вид: $ x(t) = \frac{-1}{\sqrt[4]{-t}} $.
Область определения $(-\infty; 0)$

$ x(1) = 0 \Rightarrow C = 0 $ и решение имеет вид: $ x(t) = 0 $.
Область определения $(0; +\infty)$

$ x(-1) = 0 \Rightarrow C = 0 $ и решение имеет вид: $ x(t) = 0 $.
Область определения $(-\infty; 0)$

Построим интегральные кривые:

Дабы графики не "уплывали" сильно вверх, я обрываю их на последнем значении, которое не превышает заданный $maxval$. График "реального решения" нарисован чёрным.

Также сделаем табличку, которая покажет значения полученных приближений в "точках интереса", которые заданы в условии задачи. Ниже напишем значения функции, являеющейся решением данного диф. ур., чтобы провести анализ и ответить на вопросы из следующего пункта.

```python
def euler_method(f, n, xa, xb, ya, points_of_interest, maxval = 1e9, **kw):
    #print(n, kw)
    yb = -10 ** 9
    step = (xb - xa) / n
    x = [xa]
    y = [ya]
    newx = xa
    newy = ya
    ys = []
    for i in range(n):
        newx, newy = newx + step, newy + step * f(newx, newy)
        if abs(newy) <= maxval:
            y.append(newy)
            x.append(newx)
        for j in points_of_interest:
            if abs(newx - j) <= 1e-7:
                ys.append(round(newy, 7))
        
    plt.plot(x,y, **kw)
    return ys

plt.figure(figsize=(12,15))
points = [0.25, 0.3, 0.35, 0.4]
N = [10, 30, 50, 100, 300, 500, 1000, 10000, 100000, 1000000]
cols = len(N)
color = plt.cm.rainbow(np.linspace(0, 1, cols))

print("Value in point:", end=' \t\t ')
print(*points, sep=" \t\t ")
for i,c in zip(range(cols), color):
    print("Numba of steps:", N[i], '-',
          *euler_method((lambda t, x: 2 * x * x + 2), N[i], 0, 0.5, 1, points, 600, c=c), sep='\t')

t = np.mgrid[0:0.392:eps*eps]
x = np.tan(1/4 * (8 * t + np.pi))
print('"Real function":', '-', sep='\t', end='\t')
for i in points:
    print(round(np.tan(1/4 * (8 * i + np.pi)), 7), end='\t')
plt.plot(t, x, color='0')
plt.show()
```
