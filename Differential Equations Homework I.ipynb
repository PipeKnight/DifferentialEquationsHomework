b) Решаем диф. уравнение:
$$
\frac{dx}{dt} = \frac{-2t}{5x} \\
5x dx = -2t dt \\
\int 5x dx = \int -2t dt \\
2.5 x^2 = -t^2 + C \\
x(t) = \pm \sqrt{\frac{2}{5} (C - t^2)}
$$

## Задание 2.

Приступим!

a) Строим поле направлений для уравнения:
$$ \dot x = \frac{5t}{2x} $$

```python
hide_me
plt.figure(figsize=(6,6))
plt.xlabel("$t$", fontsize=20)
plt.ylabel("$x$", fontsize=20, rotation=0)
deltax = 1
border = 10 * deltax + deltax / 2
t = np.mgrid[-border:border:deltax]
x = np.mgrid[-border:border:deltax]
ob.normdirfield(t, x, (lambda t, x: 5 * t / 2 / x), length=deltax*0.9)
```

b) Решаем диф. уравнение:
$$
\frac{dx}{dt} = \frac{5t}{2x} \\
2x dx = 5t dt \\
\int 2x dx = \int 5t dt \\
x^2 = 2.5t^2 + C \\
x(t) = \pm \sqrt{\frac{5}{2} (C + t^2)}
$$

c) Находим частные решения:

$ x(0) = 2 \Rightarrow C = 10 $ и решение имеет вид: $ x(t) = \sqrt{\frac{5}{2} (\frac{8}{5} + t^2)} $

$ x(0) = -1 \Rightarrow C = \frac{5}{2}$ и решение имеет вид: $ x(t) = -\sqrt{\frac{5}{2} (\frac{2}{5} + t^2)} $


Построим интегральные кривые:

```python
hide_me
plt.figure(figsize=(6,3))
plt.xlabel("$t$", fontsize=20)
plt.ylabel("$x$", fontsize=20, rotation=0)

deltax = 0.4
border = 10 * deltax + deltax / 2
t = np.mgrid[-border:border:deltax]
x = np.mgrid[-border * 2:border * 2:deltax]
ob.normdirfield(t, x, (lambda t, x: 5 * t / 2 / x), length=deltax*0.8)

t = np.mgrid[-border:border:0.01]
x = np.sqrt(5 / 2 * (8 / 5 + t * t))
plt.plot(t, x)
t = np.mgrid[-border:border:0.01]
x = -np.sqrt(5 / 2 * (2 / 5 + t * t))
plt.plot(t, x)
plt.show()
```
